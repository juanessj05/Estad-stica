# Task
Perform a one-way ANOVA test to compare the reaction times of three different groups. The analysis should include defining hypotheses and a significance level, preparing the data, executing the ANOVA test to calculate the F-statistic and p-value, determining the critical F-value and rejection region, visualizing the F-distribution with the calculated F-statistic and critical value, and finally, making a statistical decision and drawing practical conclusions. The ultimate goal is to present a complete summary of the ANOVA test results, including hypotheses, significance level, test statistic, p-value, statistical decision, and practical conclusion.

## Definir Hipótesis y Nivel de Significancia

### Subtask:
Establecer la hipótesis nula (H0) y la hipótesis alternativa (H1) para la prueba ANOVA de una vía y definir el nivel de significancia α = 0.05.


### Subtask: Definir Hipótesis y Nivel de Significancia

#### Hipótesis Nula (H0):
H0: Las medias de los tiempos de reacción de los tres grupos son iguales. Es decir, µ1 = µ2 = µ3, donde µ1, µ2 y µ3 son las medias poblacionales de los tiempos de reacción para el Grupo 1, Grupo 2 y Grupo 3, respectivamente.

#### Hipótesis Alternativa (H1):
H1: Al menos una de las medias de los tiempos de reacción es significativamente diferente de las otras. Es decir, al menos una de las medias (µ1, µ2, µ3) es diferente.

#### Nivel de Significancia (α):
El nivel de significancia (alpha) para la prueba es de α = 0.05. Este valor representa la probabilidad máxima de rechazar la hipótesis nula cuando es verdadera (error de Tipo I).

## Preparar los Datos

### Subtask:
Ingresar los tiempos de reacción de los tres grupos en estructuras de datos adecuadas para el análisis.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f_oneway, f

# --- 1. Definir Hipótesis y Nivel de Significancia ---
# Hipótesis Nula (H0): Las medias de las calificaciones finales de los tres métodos de enseñanza son iguales.
# H0: µA = µB = µC
# Hipótesis Alternativa (H1): Al menos una de las medias de las calificaciones finales es significativamente diferente.
# H1: Al menos una µ es diferente.
alpha = 0.01 # Nivel de significancia

print(f"--- Configuración de la Prueba ANOVA ---")
print(f"Hipótesis Nula (H0): Las medias de las calificaciones finales de los tres métodos de enseñanza son iguales (µA = µB = µC).")
print(f"Hipótesis Alternativa (H1): Al menos una de las medias de las calificaciones finales es significativamente diferente.")
print(f"Nivel de Significancia (α): {alpha}")

# --- 2. Preparar los Datos ---
metodo_a = [78, 80, 76, 82, 79, 81, 77, 80, 83, 79, 78, 80]
metodo_b = [85, 88, 84, 87, 86, 89, 90, 85, 87, 88, 86, 87]
metodo_c = [70, 72, 68, 71, 69, 73, 70, 72, 71, 70, 69, 71]

print(f"\n--- Datos de Calificaciones Finales ---")
print(f"Método A: {metodo_a}")
print(f"Método B: {metodo_b}")
print(f"Método C: {metodo_c}")

# --- 3. Realizar la Prueba ANOVA ---
f_statistic, p_value = f_oneway(metodo_a, metodo_b, metodo_c)

print(f"\n--- Resultados de la Prueba ANOVA ---")
print(f"Estadístico F: {f_statistic:.4f}")
print(f"Valor p: {p_value:.4f}")

# --- 4. Determinar el Valor Crítico de F y la Región de Rechazo ---
k = 3  # Número de grupos
n = len(metodo_a) # Número de observaciones por grupo

df1 = k - 1 # Grados de libertad del numerador
df2 = (k * n) - k # Grados de libertad del denominador

f_critical = f.ppf(1 - alpha, df1, df2) # Calcular el valor crítico de F

print(f"\n--- Determinación del Valor Crítico y Región de Rechazo ---")
print(f"Grados de Libertad del Numerador (df1): {df1}")
print(f"Grados de Libertad del Denominador (df2): {df2}")
print(f"Valor Crítico de F (alpha={alpha}): {f_critical:.4f}")
print(f"Región de Rechazo: Rechazar H0 si F-statistic > {f_critical:.4f}")

# --- 5. Graficar la Distribución F ---
x = np.linspace(0, f.ppf(0.999, df1, df2), 500) # Generar valores para el eje x de la distribución F

plt.figure(figsize=(12, 7))
plt.plot(x, f.pdf(x, df1, df2), label=f'Distribución F (df1={df1}, df2={df2})')

plt.axvline(f_statistic, color='red', linestyle='--', label=f'Estadístico F = {f_statistic:.4f}')
plt.axvline(f_critical, color='green', linestyle=':', label=f'Valor Crítico F (alpha={alpha}) = {f_critical:.4f}')

x_reject = np.linspace(f_critical, x.max(), 100)
plt.fill_between(x_reject, 0, f.pdf(x_reject, df1, df2), color='green', alpha=0.2, label='Región de Rechazo')

plt.title('Distribución F con Estadístico F y Valor Crítico')
plt.xlabel('Valor F')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# --- 6. Decisión Estadística y Conclusión Práctica ---
print(f"\n--- Decisión Estadística ---")
if f_statistic > f_critical or p_value < alpha:
    statistical_decision = "Rechazar la Hipótesis Nula (H0)"
    conclusion = (
        "Existe evidencia estadística suficiente (p < α) para afirmar que al menos una de las "
        "medias de las calificaciones finales entre los tres métodos de enseñanza es "
        "significativamente diferente. El valor p de {p_value:.4f} indica que la probabilidad de "
        "observar una diferencia tan grande o mayor entre las medias de los grupos, si la "
        "hipótesis nula fuera verdadera, es extremadamente baja ({p_value*100:.2f}%)."
    )
else:
    statistical_decision = "No rechazar la Hipótesis Nula (H0)"
    conclusion = (
        "No existe evidencia estadística suficiente (p ≥ α) para afirmar que las medias de "
        "las calificaciones finales entre los tres métodos de enseñanza son significativamente diferentes. "
        "El valor p de {p_value:.4f} indica que la probabilidad de observar una diferencia "
        "tan grande o mayor entre las medias de los grupos, si la hipótesis nula fuera verdadera, "
        "es relativamente alta ({p_value*100:.2f}%)."
    )

print(f"Decisión: {statistical_decision}")
print(f"\n--- Conclusión Práctica ---")
print(f"{conclusion}")
```

```python
from scipy.stats import f

# Definir el nivel de significancia
alpha = 0.05

# Número de grupos
k = 3

# Número de observaciones por grupo (asumiendo que todos los grupos tienen el mismo tamaño)
n = len(grupo1_tiempos)

# Grados de libertad del numerador (df1)
df1 = k - 1

# Grados de libertad del denominador (df2)
df2 = (k * n) - k

# Calcular el valor crítico de F
f_critical = f.ppf(1 - alpha, df1, df2)

print(f"Grados de Libertad del Numerador (df1): {df1}")
print(f"Grados de Libertad del Denominador (df2): {df2}")
print(f"Valor Crítico de F (alpha={alpha}): {f_critical:.4f}")
print(f"Región de Rechazo: Rechazar H0 si F-statistic > {f_critical:.4f}")
```

## Visualizar la Distribución F

### Subtask:
Generar una gráfica de la distribución F con los grados de libertad calculados, marcando el estadístico F obtenido y el valor crítico de F para visualizar la región de rechazo.

**Reasoning**:
To visualize the F-distribution and highlight the F-statistic and critical F-value, I will use `matplotlib.pyplot` and `numpy`. I will generate a range of F-values, calculate the F-distribution PDF, plot it, and then add vertical lines for the calculated F-statistic and critical F-value. Finally, I will shade the rejection region for better visualization.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Generar valores para el eje x de la distribución F
x = np.linspace(0, f.ppf(0.999, df1, df2), 500)

# Calcular la función de densidad de probabilidad (PDF) para la distribución F
plt.figure(figsize=(10, 6))
plt.plot(x, f.pdf(x, df1, df2), label=f'F-Distribution (df1={df1}, df2={df2})')

# Marcar el estadístico F calculado
plt.axvline(f_statistic, color='red', linestyle='--', label=f'F-Statistic = {f_statistic:.4f}')

# Marcar el valor crítico de F
plt.axvline(f_critical, color='green', linestyle=':', label=f'Critical F-Value (alpha=0.05) = {f_critical:.4f}')

# Sombreado de la región de rechazo
x_reject = np.linspace(f_critical, x.max(), 100)
plt.fill_between(x_reject, 0, f.pdf(x_reject, df1, df2), color='green', alpha=0.2, label='Rejection Region')

plt.title('F-Distribution con Estadístico F y Valor Crítico')
plt.xlabel('F-Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
